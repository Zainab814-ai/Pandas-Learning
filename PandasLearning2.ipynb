```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
df=pd.DataFrame(housing.data,columns=housing.feature_names)
print(df)
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df.isnull()
```

```python
df.columns
```

```python
df["HouseAge"]=df["HouseAge"].astype(int)
df.head()
```

```python
df["price_in_lacs"]=df["HouseAge"]/10000
df.head()
```

```python
df=df[df["HouseAge"]<40]
df.head()
```

```python
stats=df.groupby("HouseAge").agg({
    "AveRooms":["count","sum"],
    "Population":["sum"]
})
print(stats.head(7))
stats.columns=[
    "Totalrooms",
    "Sumofrooms",
    "totalPopulation"
]
stats=stats.reset_index()
print(stats)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ali Khan", "Sara Ahmed", "Zain Ali", "Hassan Khan", None],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi"],
    "Email": ["ali@gmail.com", "sara@yahoo.com", None, "hassan@hotmail.com", "zain@outlook.com"],
    "Gender": ["Male", "Female", "Male", None, "Male"]
})
print(df)
```

```python
df.isnull()
```

```python
df.duplicated()
```

```python
df[["Email","Gender","Name"]]=df[["Email","Gender","Name"]].fillna('Unknown')
print(df)
```

```python
df["Name"].str.upper()
```

```python
df[df["Name"].str.upper().str.startswith("SARA")]
```

```python
df[df["Name"].str.upper().str.endswith("KHAN")]
```

```python
df["Gender"]=df["Gender"].replace({"Male":"M","Female":"F"})
print(df)
```

```python
df["Username"]=df["Email"].str.split("@").str[0]
print(df)
df["Domain"]=df["Email"].str.split("@").str[1]
print(df)
```

```python
df["City"]=pd.to_numeric(df["City"],errors="coerce")
print(df)
```
